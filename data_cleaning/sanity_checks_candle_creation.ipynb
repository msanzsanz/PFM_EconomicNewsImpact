Ok. I created my candles with a window size of 15 and 30 minutes, starting from 00:00. I.e.

00:00 - 00:15
00:15 - 00:30
00:30 - 00:45 
...

As some news are published in between those periods (e.g. 09:55), the join with the pairs dataframe does not work for them (09:55 + 15 min = 10:10, so there is no match with any candle).

A potential solution could be to create candles individually per each new, creating 15/30 min candles starting when each new was published, not starting by 00:00 for the entire dataset.

However, does it really worth the effort for the baseline model? Perhaps not, as remeber why we are creating this candles:

- 5 min candles:  We need fine granularity here, as the hypotesis is that market reaction right after the publication of the new is key for the models. 
There are not nan here, as we rounded the publication of the new to the closest 5min slot.

- 15 / 30 min candles: Before the new publication, this will be used to know whether the market was flat or was having a high volatility. Thus, we can apply the same criteria to round to the closest candle.

After the publication, this features will be used to know whether the new caused a positive or negative reaction in the market. Thus, I think we can apply the same criteria to round to the closest candle.


Therefore, let´s just round to the closest candle in the code for now.



Before rounding the df up, let´s review how many / which news are affected by this.
