```python
import pandas as pd
```

```python
df = pd.read_csv('features_news_USD_pair_EURUSD_2007_2018.csv')
```

```python
df_news = df[['datetime', 'datetime_gmt',
 'actual',
 'forecast',
 'forecast_error',
 'impact',
 'new',
 'previous',
 'previous_error',
 'forecast_error_diff',
 'weekday',
 'previous_value',
 'previous_error_diff',
 'total_error_diff',
 'forecast_error_diff_outlier_class',
 'forecast_error_diff_deviation',
 'previous_error_diff_outlier_class',
 'previous_error_diff_deviation',
 'total_error_diff_outlier_class',
 'total_error_diff_deviation']].copy()
```

```python
len(df_news)
```

```python
df_market = df[['datetime',
 'pips_candle_0_30_after',
 'pips_candle_max_0_30_after',
 'pips_candle_min_0_30_after',
 'pips_candle_0_60_after',
 'pips_candle_max_0_60_after',
 'pips_candle_min_0_60_after',
 'pips_candle_0_120_after',
 'pips_candle_max_0_120_after',
 'pips_candle_min_0_120_after',
 'pips_candle_0_180_after',
 'pips_candle_max_0_180_after',
 'pips_candle_min_0_180_after',
 'pips_candle_0_240_after',
 'pips_candle_max_0_240_after',
 'pips_candle_min_0_240_after',
 ]]
```

```python
df_market.head(2)
```

```python
df_market = df_market.drop_duplicates()
```

```python
df_market.columns
```

```python
df_market.columns = ['datetime', 'pips_candle_0_30',
       'pips_candle_max_0_30', 'pips_candle_min_0_30',
       'pips_candle_0_60', 'pips_candle_max_0_60',
       'pips_candle_min_0_60', 'pips_candle_0_120',
       'pips_candle_max_0_120', 'pips_candle_min_0_120',
       'pips_candle_0_180', 
       'pips_candle_max_0_180', 'pips_candle_min_0_180',
       'pips_candle_0_240', 
       'pips_candle_max_0_240', 'pips_candle_min_0_240']
```

```python
df_market_long = pd.wide_to_long(df_market, stubnames=['pips_candle_0_','pips_candle_max_0_', 'pips_candle_min_0_'], i=['datetime'], j='window')
```

```python
df_market_long.head(10)
```

```python
df_market_long = df_market_long.reset_index()
```

```python
df_market_long.head(2)
```

```python
df_news = df_news.merge(df_market_long, on='datetime', how='left')
```
