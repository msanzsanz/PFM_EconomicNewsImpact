We want to create a new dataframe for enabling the following visualization:

- Sort which are the most secure news to invest because of their predictability. I.e. positive (actual-forecast) deviations always move the market in favour of the USD and the other way around.

```python
import numpy as np
df = pd.read_csv('features_news_USD_pair_EURUSD_2007_2018.csv')
```

```python
columns_of_interest = [
 'datetime',
 'forecast_error',
 'impact',
 'new',
 'previous_error',
 'pips_candle_max_0_30_after',
 'pips_candle_min_0_30_after',
 'pips_candle_max_0_60_after',
 'pips_candle_min_0_60_after',
 'pips_candle_max_0_120_after',
 'pips_candle_min_0_120_after',
 'pips_candle_max_0_180_after',
 'pips_candle_min_0_180_after',
 'pips_candle_max_0_240_after',
 'pips_candle_min_0_240_after',
]
```

```python
df_temp = df[columns_of_interest]
```

```python
df_temp.head(3)
```

```python
df_temp['forecast_error'] = df_temp['forecast_error'].apply(lambda x: np.where(x == 'accurate', 0, 
                                                                              np.where(x == 'better', 1, -1)))
df_temp['previous_error'] = df_temp['previous_error'].apply(lambda x: np.where(x == 'accurate', 0, 
                                                                              np.where(x == 'better', 1, -1)))
df_temp['impact'] = df_temp['impact'].apply(lambda x: np.where(x == 'Low', 1, 
                                                                              np.where(x == 'Medium', 2, 3)))
df_temp['forecast_error'] = df_temp['forecast_error'] * df_temp['impact']
df_temp['previous_error'] = df_temp['previous_error'] * df_temp['impact']
df_temp['total_error'] = df_temp['forecast_error']  + df_temp['previous_error'] 
```

```python
df_temp.head(3)
```

```python
df_all_news = df_temp.groupby('datetime').agg({'new': lambda x: list(x),
                                               'total_error': 'sum',
                                               'pips_candle_max_0_30_after': 'first',
                                               'pips_candle_max_0_60_after': 'first',
                                               'pips_candle_max_0_120_after': 'first',
                                               'pips_candle_max_0_180_after': 'first',
                                               'pips_candle_max_0_240_after': 'first',
                                                   }).reset_index()
df_all_news['new'] = df_all_news['new'].apply(lambda x: '_'.join(x))
```

```python
df_all_news.head(2)
```

```python
df_all_news = df_all_news.reset_index()
```

```python
def get_predictable_flag(row):
    error = row['total_error']
    if error > 0 and row['pips_candle_max_0_30_after'] > 15:
        return 1
    else:
        return 0
```

```python
df_all_news['predictable'] = df_all_news.apply(lambda row:get_predictable_flag(row))
```

```python
df_all_news.to_csv('predictability_all.csv')
```

```python
df_temp = df[columns_of_interest]
df_temp = df_temp[df_temp['impact'] == 'High']
df_temp['forecast_error'] = df_temp['forecast_error'].apply(lambda x: np.where(x == 'accurate', 0, 
                                                                              np.where(x == 'better', 1, -1)))
df_temp['previous_error'] = df_temp['previous_error'].apply(lambda x: np.where(x == 'accurate', 0, 
                                                                              np.where(x == 'better', 1, -1)))
df_temp['impact'] = df_temp['impact'].apply(lambda x: np.where(x == 'Low', 1, 
                                                                              np.where(x == 'Medium', 2, 3)))
df_temp['forecast_error'] = df_temp['forecast_error'] * df_temp['impact']
df_temp['previous_error'] = df_temp['previous_error'] * df_temp['impact']
df_temp['total_error'] = df_temp['forecast_error']  + df_temp['previous_error'] 

df_all_news = df_temp.groupby('datetime').agg({'new': lambda x: list(x),
                                               'total_error': 'sum',
                                               'pips_candle_max_0_30_after': 'first',
                                               'pips_candle_max_0_60_after': 'first',
                                               'pips_candle_max_0_120_after': 'first',
                                               'pips_candle_max_0_180_after': 'first',
                                               'pips_candle_max_0_240_after': 'first',
                                                   }).reset_index()
df_all_news['new'] = df_all_news['new'].apply(lambda x: '_'.join(x))
```

```python
df_all_news.to_csv('predictability_high.csv')
```
